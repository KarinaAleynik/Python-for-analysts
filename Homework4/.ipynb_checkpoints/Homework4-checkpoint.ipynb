```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Задача 1

### 1.1. Скачать данные

```python
df = pd.read_csv('kc_house_data.csv', encoding = 'windows-1251', parse_dates=['date'])
```

```python
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(20, 20))

sns.histplot(df['price'], bins=85)

plt.title('Стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Кол-во домов');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(20, 10))

sns.histplot(df['sqft_living'], bins=95)

plt.title('Распределение квадратуры жилой')
plt.xlabel('Жилая квадратура')
plt.ylabel('Кол-во домой');
```

+-75% домов имеют жилую квадратуру от 1500 до 2500 кв.м.

### 2.4 Изучите распределение года постройки

```python
plt.figure(figsize=(20, 10))

sns.histplot(df['yr_built'], bins=95)

plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Кол-во домов');
```

Глядя на график, можно сделать следующие выводы:

- Строительство домов из данного списка началось в 1900 году.
- Спад строительства наблюдался в 1930-х.
- Пик строительства приходится на начало 2000-х.
- Прослеживается тенденция на строительство домов

## Задача 2

### 2.1 Изучите распределение домов от наличия вида на набережную
Постройте график,сделайте выводы

```python
counts = df['waterfront'].value_counts()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(counts, autopct='%1.1f%%')

plt.title('Распределение домов от наличия вида на набережную')
plt.legend(['Без вида', 'С видом']);
```

Значительное большинство домов (0.8 %) не имеет вида на набережную

### 2.2 Изучите распределение этажей домов

```python
counts_floors = df['floors'].value_counts()
    
plt.figure(figsize=(8, 6))    
plt.title('Распределение этажей домов')
sns.barplot(x=counts_floors.index, y=counts_floors.values)
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во домов')
plt.show()
```

Около половины домов являются одноэтажными. Порядка 40 % приходится на двухэтажные дома

### 2.3 Изучите распределение состояния домов

```python
counts_grade = df['condition'].value_counts()
    
plt.figure(figsize=(8, 6))    
plt.title('Распределение состояния домов')
sns.barplot(x=counts_grade.index, y=counts_grade.values)
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во домов')
plt.show()
```

Большинство домов находятся в "удовлетворительном" состоянии и лучше

## Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Проверим как зависит цена от ключевых параметров таких как: 
- кол-во спален;
- кол-во ванных комнат;
- площадь жилых кв. метров;
- кол-во этажей;
- состояние дома;
- год постройки;
- качество конструкции и дизайна.

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg');
```

Вывод: стоимость умеренно зависит от количества спален в доме.

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg');
```

Вывод: стоимость заметно зависит от количества ванных комнат в доме. 

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

Вывод: стоимость сильно зависит от жилых кв.метров в доме.

```python
sns.jointplot(x=df['price'], y=df['floors'], kind='reg');
```

Стоимость дома не зависит от кол-ва этажей в доме

```python
sns.jointplot(x=df['price'], y=df['condition'], kind='reg');
```

Стоимость дома не зависит от состояния дома

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
```

Стоимость не зависит от года постройки

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

Прослеживается сильная зависимость между стоимостью и качеством конструкции и дизайна

Из графиков следует, что:
- сильнее всего стоимость зависит от площади жилых кв. метров и от качества конструкции и дизайна;
- нет зависимости между стоимостью и годом постройки дома.

В подтверждении данного вывода проведем корреляционный анализ.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

И теперь на тепловой карте посмотрим наличие линейных связей

```python
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.0)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Действительно, наблюдается взаимосвязь между стоимостью и площадью жилых кв. метров, качества конструкции и дизайна дома. Она равна 0.7, и ее можно классифицировать, как "сильную". Остальные же параметры показвают "слабую" или "среднюю" линейную связь.
